# Exploring the difference between each decade of basketball

Hello, today we're going to be exploring what has differed within each decade of basketball and some other statistics that we can derive from this dataset

Firstly, i'm going to need to import all of our python dependencies and then our datasets so we can get to work

```python
import pandas as pd
import numpy as np
import seaborn as sbs
```

```python
player_stats
```

```python
players_2010 = player_stats[player_stats['Year'] >= 2010]
```

```python
players_2000 = player_stats[(player_stats['Year'] >= 2000) & (player_stats['Year'] < 2010)]
```

```python
players_1990 = player_stats[(player_stats['Year'] >= 1990) & (player_stats['Year'] < 2000)]
```

```python
players_1980 = player_stats[(player_stats['Year'] >= 1980) & (player_stats['Year'] < 1990)]
```

```python
# All the years to separate
years_to_get = [
    [players_2010, 2010, 8],
    [players_2000, 2000, 10],
    [players_1990, 1990, 10],
    [players_1980, 1980, 10]
]

# This holds all dataframes for each decade and year WITHIN that decade
output_df = []

# Iterate over the all the unfilterd data frames of each decade and create unique players only
for i in range(len(years_to_get)):
    years = []
    current_decade = years_to_get[i][0]
    
    # Iterate over the total amt of seasons within each decade
    for j in range(years_to_get[i][2]):
        current_year = years_to_get[i][1] + j
        current_df = current_decade[current_decade.Year == current_year]
        years.append(current_df.drop_duplicates(subset='Player', keep='first'))
     
    output_df.append(years)

# Create all the unique player dataframes (once per season)
u_players_2010 = pd.concat(output_df[0])
u_players_2000 = pd.concat(output_df[1])
u_players_1990 = pd.concat(output_df[2])
u_players_1980 = pd.concat(output_df[3])
u_players_2010
```

Now that we have all the dataframes divided by the decades players had played in, let's find out how many games on average players were playing per season

```python
gp_2010 = u_players_2010.G[u_players_2010.G.notnull()].mean()
gp_2000 = u_players_2000.G[u_players_2000.G.notnull()].mean()
gp_1990 = u_players_1990.G[u_players_1990.G.notnull()].mean()
gp_1980 = u_players_1980.G[u_players_1980.G.notnull()].mean()
```

```python
gp_data = [
    ["1980", gp_1980],
    ["1990", gp_1990],
    ["2000", gp_2000],
    ["2010", gp_2010]
]
gp_df = pd.DataFrame(columns=["Decade", "games played"], data=gp_data)
```

```python
sbs.barplot(x='Decade', y='games played', data=gp_df)
```

```python
sbs.lineplot(x='Decade', y='games played', data=gp_df)
```
